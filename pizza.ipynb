```python
daily_sales = pizza_store.groupby('order_date')['quantity'].sum()
print(daily_sales)
```

```python
import matplotlib.pyplot as plt
daily_sales.plot(kind='line', figsize=(10, 5), )
plt.title('daily_sales')
plt.xlabel('order_date')
plt.ylabel('sales')
plt.show
```

```python
Ttype_pizza = pizza_store.groupby('pizza_name')['quantity'].sum()
print(Ttype_pizza)
```

```python
import matplotlib.pyplot as plt

Ttype_pizza.plot(kind='bar',figsize=(10, 5))
Ttype_pizza.index = Ttype_pizza.index.str[:5]
plt.title('total_pizza_type')
plt.xlabel("pizza_type")
plt.ylabel("quantity")
plt.show
```

```python
best_selpizza = pizza_store.groupby('pizza_name')['quantity'].sum().nlargest(5)
print(best_selpizza)
```

```python
best_selpizza.plot(kind='bar',figsize=(10, 5))
best_selpizza.index = best_selpizza.index.str[:10]
plt.title('total_pizza_type')
plt.xlabel("pizza_type")
plt.ylabel("quantity")
plt.show
```
